# Plot Risk Maps

This IPython scripts allows users to plot risk maps from scenario or probabilistic risk analysis. If more than an asset per location exists, it is possible to plot the aggregated loss, or create various maps according to each building taxonomy. For the latter option, it is necessary to specify also the location of the exposure and vulnerability models used in the calculations.
<br />
<br />
Please specify the type of plotting in variable ploting_type as follows:  <br />
0 - Aggregated loss map only <br />
1 - Loss maps per taxonomy only <br />
2 - Both aggregated and taxonomy-based <br /> 
<br />
Please specify the limits of bounding box for the map in variable bounding_box. If you would like to use the limits from the loss map, set this variable to 0.

```python
import plot_risk_maps as plotter

plotting_type = 2
risk_map = 'loss-maps-258.xml'
bounding_box = 0
marker_size = 50
log_scale = True
export_map_to_csv = True
exposure_model = 'exposure_model2011_MAL.xml'

plotter.build_map(plotting_type,risk_map,bounding_box,log_scale,exposure_model,marker_size,export_map_to_csv)
```
